```python
float_able = []
float_unable = set(["Unnamed: 0", 'GET',
 'POST',
 'TRACE',
 'HTTP/1.0',
 'HTTP/1.1',
 '0',
 ])
for c in df.columns:
    if c in float_unable: continue
    try:
        try:
            if df[c].str.startswith("0x").astype("int").sum():
                df[c] = pd.Series(int(i,16) for i in df[c])
        except: pass
        df[c].astype("float")
    except Exception as e:
        # print(c, e) 
        float_unable.add(c)
print(all([(c in float_unable) for c in df.columns]))
float_able = [c for c in df.columns if c not in float_unable]
```

```python
temp = df.copy(True)
correlation = temp[[(c) for c in float_able if c not in ["ip.src_host", "ip.dst_host", "http"]]].corr("pearson")
```

```python
correlation
```

```python
correlation_crudential = set()
for col1 in correlation.columns:
    for col2 in correlation.columns:
        if col1 == col2: continue
        if correlation[col1][list(correlation.index).index(col2)] >= .2:
            correlation_crudential.add((col1, col2, correlation[col1][list(correlation.index).index(col2)]))
```

```python
"""
arp.opcode : arp Operation Code : 요청이나 응답 패킷인지와 발생하는 주소 해결 유형을 정의
arp.hw.size : Hardware Address Length : 해당 패킷에 사용되는 물리 주소의 길이(byte)를 정의

tcp.len 
tcp.flags
tcp.flags.ack
tcp.seq
tcp.dstport

http.content_length : http response의 길이 (크기)
http.response : http response가 올바르게 도착하였는가 ( status == 200 을 의미)
"""
_ = [(a,b,c) for a,b,c in correlation_crudential if c >= .5]
_[:len(_)//2]
```

```python
available_columns = set()
for c1, c2, val in list(correlation_crudential):
    available_columns.add(c1)
    available_columns.add(c2)
```

```python
available_columns # one by one correlation
```

```python
new_df = pd.DataFrame()
for c1 in float_able:
    for c2 in float_able:
        if c1 == c2: continue
        if f"{c2}-{c1}" in new_df.columns: continue
        new_df[f"{c1}-{c2}"] = df[c1] + df[c2]
```

```python
correlation = new_df.corr("pearson")
```

```python
correlation_crudential = set()
for col1 in correlation.columns:
    for col2 in correlation.columns:
        if col1 == col2: continue
        if correlation[col1][list(correlation.index).index(col2)] >= .2:
            correlation_crudential.add((col1, col2, correlation[col1][list(correlation.index).index(col2)]))
```

```python
available_columns = set()
for c1, c2, val in list(correlation_crudential):
    available_columns.add(c1)
    available_columns.add(c2)

available_columns_copy = set()
for element in available_columns:
    for c in element.split("-"):
        available_columns_copy.add(c)
available_columns = available_columns_copy
del available_columns_copy
```
